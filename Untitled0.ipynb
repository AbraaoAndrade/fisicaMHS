```
import matplotlib.pyplot as plt
import numpy as np

a = 2
w = 2
o = 2
f = 0
p = 0
pi = 3.141592654

t = np.arange(0.0, 10.0, 0.01)
x = a * np.cos((w * t)+o)
v = (-1) * a * w * np.sin((w * t)+o)
a = (-1) * a * w * w * np.cos((w * t)+o)

fig, ax = plt.subplots()
ax.plot(t, x)
ax.plot(t, v)
ax.plot(t, a)

ax.set(xlabel='tempo (s)', ylabel='posição (m) - velocidade (m/s) - aceleração (m/s²)',
       title='MHS')
ax.grid()

fig.savefig("test.png")
plt.show()
```
