```
import matplotlib.pyplot as plt
import numpy as np

print('os valores que não conhecer digite "0"')

x0 = 0
v0 = 0

pi = 3.141592654
A = float(input('digite a amplitude: ')) #0.25
if A == 0:
  x0 = float(input('digite o deslocamento inicial: ')) # 0.015
  v0 = float(input('digite a velocidade incial: ')) # 0.4
  
w = float(input('digite a velocidade angular: ')) #20
if w == 0:
  f = float(input('digite a frequencia: ')) #3.2
  if f == 0:
    T = float(input('digite o periodo: ')) #0.31
  if T == 0:
    k = float(input('digite a constante da mola: ')) # 200
    m = float(input('digite a massa do corpo que oscila: ')) #0.5 
    
o = float(input('digite o anfulo de fase: ')) #-0.93
if o == 0 and v0 == 0 and x0 == 0:
  x0 = float(input('digite o deslocamento inicial: ')) # 0.015
  v0 = float(input('digite a velocidade incial: ')) # 0.4


if w == 0:
  if k != 0 and m != 0: 
    w = (k / m) ** (1/2)
  elif f != 0:
    w = 2*pi*f
  elif T != 0:
    w = (2*pi)/T
    
if A == 0 and x0 != 0 and v0 != 0 and w != 0:
  A = ((x0 ** 2) + ((v0 ** 2) / (w ** 2))) ** (1/2)
  
if o == 0 and x0 != 0 and v0 != 0 and w != 0:
  o = np.arctan(-(v0/(w*x0)))

if A == 0 or w == 0 or o == 0:
  print("variaveis faltando")
  
xmax = A
vmax = w * A
amax = w * w * A

print('\n')

print('Xmax :' , xmax)
print('Vmax :' , vmax)
print('Amax :' , amax)

t = np.arange(0.0, 0.8, 0.01)
p = A * np.cos((w * t)+o)
v = (-1) * A * w * np.sin((w * t)+o)
ac = (-1) * A * w * w * np.cos((w * t)+o)

#posicão
fig, ax = plt.subplots()
ax.plot(t, p)

ax.set(xlabel='tempo (s)', ylabel='posição (m)', title='MHS - posição')
ax.grid()

fig.savefig("test.png")
plt.show()

#velocidade
fig, ax = plt.subplots()
ax.plot(t, v)

ax.set(xlabel='tempo (s)', ylabel='velocidade (m/s)', title='MHS - velocidade')
ax.grid()

fig.savefig("test.png")
plt.show()

#aceleração
fig, ax = plt.subplots()
ax.plot(t, ac)

ax.set(xlabel='tempo (s)', ylabel='aceleração (m/s²)', title='MHS - aceleração')
ax.grid()

fig.savefig("test.png")
plt.show()
```
